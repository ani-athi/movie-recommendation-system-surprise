### Merging the dataset

```python
# Merging
netflix_merged = pd.merge(netflix, title, on='movie_id', how ='left')
netflix_merged.head()
```

# 1. Finding the list of most popular and liked genre

```python
popular_genre = (netflix_merged.groupby('genres')['rating']
                 .count())
```

```python
popular_genre.sort_values(ascending=False)
```

```python
liked_genre = (netflix_merged.groupby('genres')['rating']
               .mean()
               .sort_values(ascending = False)
               .reset_index()
               .rename(columns={'rating': 'avg_rating'})
)

liked_genre.head(10)
```

# 2. Finding what genre movies have received the best and worst ratings based on User Rating

```python
best_genre = liked_genre.loc[liked_genre['avg_rating'].idxmax()]
worst_genre = liked_genre.loc[liked_genre['avg_rating'].idxmin()]

print(f"Best Genre: {best_genre['genres']}")
print(f"Worst Genre: {worst_genre['genres']}")
```

```python
trainset = data.build_full_trainset()
model.fit(trainset)
```
